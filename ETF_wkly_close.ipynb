```python
import requests
import pandas as pd
import json
```

```python
#Pull SPY weekly close price api data

url = "https://mboum-finance.p.rapidapi.com/hi/history"

querystring = {"symbol":"SPY","interval":"1wk","diffandsplits":"false"}

headers = {
	"X-RapidAPI-Key": "",
	"X-RapidAPI-Host": "mboum-finance.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)

print(response.json())
```

```python
# make json output somewhat readable
data = json.loads(response.text)

print(json.dumps(data,indent=1))
```

```python
#convert json response to dataframe
spy_data=response.json()
spy_df=pd.json_normalize(spy_data['items'].values())
spy_df
```

```python
#reset index to date
spy_df=spy_df.set_index('date')
spy_df
```

```python
#drop columns except for close
spy_df.drop(spy_df.columns[[0,1,2,3,5,6]],axis=1,inplace=True)
spy_df
```

```python
#rename close column to etf ticker name 
spy_df.rename(columns={'close':'SPY wkly close'},inplace=True)
spy_df
```

```python
#get QQQ, DIA, VNQ, and BITO etf weekly close price via api

url = "https://mboum-finance.p.rapidapi.com/hi/history"

querystring = {"symbol":"QQQ","interval":"1wk","diffandsplits":"false"}

headers = {
	"X-RapidAPI-Key": "",
	"X-RapidAPI-Host": "mboum-finance.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)


#formating response into dataframe with wkly close price

qqq_data=response.json()
qqq_df=pd.json_normalize(qqq_data['items'].values())

qqq_df=qqq_df.set_index('date')

qqq_df.drop(qqq_df.columns[[0,1,2,3,5,6]],axis=1,inplace=True)

qqq_df.rename(columns={'close':'QQQ wkly close'},inplace=True)
qqq_df
```

```python
#merge all wkly etf close price

wkly_close_etf=pd.concat([spy_df,qqq_df,dia_df,vnq_df],axis=1)

wkly_close_etf
```

```python
#Project 2 finance starter portfolio

import pandas as pd
import numpy as np
```

```python
#Historical US Treasury Bond Rates from 01/31/2000 - 08/31/2023

Tbond_rates=pd.read_csv('../Project-2/Resources/FRB_H15.csv',index_col = "Series Description", infer_datetime_format = True, parse_dates = True).sort_index(ascending=True)

pd.set_option('display.max_columns', None)

Tbond_rates
```

```python
#identify dtypes
Tbond_rates.dtypes
```

```python
#cleaning dataframe removing rows that aren't dates

Tbond_rates.drop(Tbond_rates.index[284:289],inplace=True)

Tbond_rates
```

```python
#change index name
Tbond_rates=Tbond_rates.rename_axis('Date')
Tbond_rates
```

```python
#drop commercial paper interest rate and market rate discount basis 
Tbond_rates.drop(Tbond_rates.iloc[:,1:13],inplace=True,axis=1)
Tbond_rates
```

```python
#drop bonds held to maturity,quoted on investment basis, inflation-indexed
Tbond_rates.drop(Tbond_rates.iloc[:,12:17],inplace=True,axis=1)
Tbond_rates
```

```python
#find null values(no rates)
Tbond_rates.isnull().sum().sum()
```

```python
#obtain yields from last 20 years
yields_last_20years=Tbond_rates.loc['2003-01':'2023-08']
yields_last_20years
```

```python
yields_last_20years.dtypes
```

```python
yields_last_20years=yields_last_20years.astype(float)
```

```python
yields_last_20years.dtypes
```

```python
#obtain yields 1 year history
yields_last_year=yields_last_20years.loc['2022-08':'2023-08']
yields_last_year
```

```python
yields_last_year.mean()
```

```python
yields_last_year.to_csv('US_bond_yields.csv')
```
